```python
%matplotlib notebook
%precision 3

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

import matplotlib
import math
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [9.5, 7]
import datetime
import pandas
from matplotlib_venn import venn3
from matplotlib_venn import venn2

```

```python
prefix='/home/ahu/git/antonie'
data=pandas.read_csv(prefix+"/combined-codons-s-protein.csv")
data.head()
data=data.reset_index()
```

```python
def amalg(row, nam):
    return str(row["index"])+" "+row[nam]

data["indorig"] = data.apply(lambda row: amalg(row, "orig"), axis=1)
data["indcvac"] = data.apply(lambda row: amalg(row, "cvac"), axis=1)        
data["indbntc"] = data.apply(lambda row: amalg(row, "bntc"), axis=1)        

```
